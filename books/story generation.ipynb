```python
import os
from langchain import PromptTemplate, LLMChain
from langchain.llms import OpenAI

# Set OpenAI API key
os.environ["OPENAI_API_KEY"] = "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"

# Create templates for different parts
summary_template = """Create a brief story outline about {topic}.
Include the main plot points and potential chapter breakdown.
Format the output as:

Title: [Story Title]
Summary: [2-3 sentences describing the overall story]
Chapter Outlines:
1. [Brief description of chapter 1]
2. [Brief description of chapter 2]
3. [Brief description of chapter 3]
4. [Brief description of chapter 4]
5. [Brief description of chapter 5]
"""

chapter_template = """Write chapter {chapter_num} for a story with the following context:
Title: {title}
Summary: {summary}
Chapter Outline: {chapter_outline}

Additional Instructions: {feedback}

Format as:
Chapter {chapter_num}:
Title: [Chapter Title]
Text: [Chapter text, approximately 200 words]
"""

# Set up LLM
llm = OpenAI(temperature=0.7, model="gpt-3.5-turbo-instruct", max_tokens=2000)

def get_user_feedback():
    print("\nPlease provide your feedback:")
    print("1. Continue to next chapter")
    print("2. Regenerate this chapter with changes")
    print("3. Exit story generation")
    choice = input("Enter your choice (1/2/3): ")
    
    if choice == "2":
        feedback = input("What changes would you like to see? ")
        return choice, feedback
    return choice, ""

def generate_chapter(chapter_num, title, summary, chapter_outline, feedback=""):
    chapter_prompt = PromptTemplate(
        template=chapter_template,
        input_variables=["chapter_num", "title", "summary", "chapter_outline", "feedback"]
    )
    chapter_chain = LLMChain(prompt=chapter_prompt, llm=llm)
    return chapter_chain.run(
        chapter_num=chapter_num,
        title=title,
        summary=summary,
        chapter_outline=chapter_outline,
        feedback=feedback
    )

def main():
    # Get initial topic
    topic = input("Enter your story topic: ")
    
    # Generate story summary
    summary_prompt = PromptTemplate(template=summary_template, input_variables=["topic"])
    summary_chain = LLMChain(prompt=summary_prompt, llm=llm)
    story_outline = summary_chain.run(topic)
    
    print("\n=== Story Outline ===")
    print(story_outline)
    
    # Get feedback on summary
    print("\nIs this story outline acceptable?")
    proceed = input("Would you like to proceed with this outline? (yes/no): ").lower()
    
    if proceed != 'yes':
        print("Let's start over!")
        return
    
    # Parse the outline
    lines = story_outline.split('\n')
    title = next(line.split(': ')[1] for line in lines if line.startswith('Title:'))
    summary = next(line.split(': ')[1] for line in lines if line.startswith('Summary:'))
    chapter_outlines = []
    capture_outlines = False
    
    for line in lines:
        if line.startswith('Chapter Outlines:'):
            capture_outlines = True
            continue
        if capture_outlines and line.strip() and line[0].isdigit():
            chapter_outlines.append(line[3:])  # Remove the number and dot
    
    # Generate chapters one by one with feedback
    final_story = [story_outline]
    
    for chapter_num in range(1, 6):
        while True:
            print(f"\n=== Generating Chapter {chapter_num} ===")
            chapter_content = generate_chapter(
                chapter_num,
                title,
                summary,
                chapter_outlines[chapter_num-1]
            )
            print(chapter_content)
            
            choice, feedback = get_user_feedback()
            
            if choice == "1":  # Continue to next chapter
                final_story.append(chapter_content)
                break
            elif choice == "2":  # Regenerate with feedback
                continue
            else:  # Exit
                print("Stopping story generation.")
                return
    
    # Save complete story
    with open("story.txt", 'w+') as file:
        file.write("\n\n".join(final_story))
    
    print("\nStory has been completed and saved to 'story.txt'!")

if __name__ == "__main__":
    main()
```
